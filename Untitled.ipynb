```python
#Get Price and Name of Product
def get_name_price(product_url):
    

    price = driver.find_element(By.CLASS_NAME,"original").text
    name  = driver.find_element(By.CLASS_NAME, "product-intro__head-name").text
    print("Name: "+str(name))
    print("price: "+ str(price))
    return name,price
```

```python
#GET Images and download them
#Rename images with the same name of the product
product_url="https://roe.shein.com/SHEIN-Puff-Sleeve-Schiffy-Tee-p-10304672-cat-1738.html?scici=productDetail~~RecommendList~~RS_emarsys,RJ_NoFaultTolerant~~Customers%20Also%20Viewed~~SPcProductDetailCustomersAlsoViewedList~~0&mallCode=1"
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get(product_url);
driver.maximize_window()
name,price = get_name_price("https://roe.shein.com/SHEIN-Puff-Sleeve-Schiffy-Tee-p-10304672-cat-1738.html?scici=productDetail~~RecommendList~~RS_emarsys,RJ_NoFaultTolerant~~Customers%20Also%20Viewed~~SPcProductDetailCustomersAlsoViewedList~~0&mallCode=1")
imgs = driver.find_elements(By.TAG_NAME,"img")
counter=0
dirt="C:\\Users\slugf\\Downloads\\"+name.replace(" ","-")
if not os.path.isdir(dirt):

    os.mkdir(dirt)
    for img in imgs:
        if img.get_attribute("aria-label")==name:
            counter=counter+1
            print(img.get_attribute("data-src"))
            img_url="https:"+str(img.get_attribute("data-src"))
            urllib.request.urlretrieve(img_url, dirt+"\\"+name.replace(" ","-")+str(counter)+".png")

    #driver.close()
```
